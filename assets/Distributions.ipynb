```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_context('poster')
sns.set_style('ticks')
```

```python
age_mean = 35.3
age_sd = 6.8
age = np.random.normal(age_mean, age_sd, 10000)
_ = sns.distplot(age,color='#ff00b7', hist=False, kde_kws={"shade": True})
sns.despine()
figure = _.get_figure()
figure.savefig('age.svg')
```

```python
experience_mean = 14.0
experience_sd = 5.8
experience = np.random.normal(experience_mean, experience_sd, 10000)
_ = sns.distplot(experience, color='#00ff63', hist=False, kde_kws={"shade": True})
sns.despine()
figure = _.get_figure()
figure.savefig('experience.svg')
```
